```python
import os
import pandas as pd
from pathlib import Path
import hvplot.pandas
import requests
import yfinance as yf
from finta import TA
%matplotlib inline
```

```python
market_data = yf.download("SPY", start="2017-01-01", end="2022-10-19",interval = "1d")
```

```python
display(market_data.head(5))
display(market_data.tail(5))
```

```python
market_data.drop(columns=["Adj Close"],inplace=True)
```

```python
display(market_data.head(5))
display(market_data.tail(5))
```

```python
bbands_df = TA.BBANDS(market_data)
```

```python
bbands_df
```

```python
market_data_boil = pd.concat([market_data, bbands_df], axis=1)
```

```python
display(market_data_boil.head(5))
display(market_data_boil.tail(5))
```

```python
# Set the short window, medium, long windows.
short_window = 20
medium_window = 50
long_window = 200

# Add the SMA technical indicators for the short, medium and long windows
market_data_boil["SMA - 20"] = TA.SMA(market_data_boil, short_window)
market_data_boil["SMA - 50"] = TA.SMA(market_data_boil, medium_window)
market_data_boil["SMA - 200"] = TA.SMA(market_data_boil, long_window)
```

```python
display(market_data_boil.head(5))
display(market_data_boil.tail(5))
```

```python
market_data_boil.to_csv('market_data.csv')
```
